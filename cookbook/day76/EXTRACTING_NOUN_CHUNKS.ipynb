# EXTRACTING NOUN CHUNKS

Noun chunks are known in linguistics as noun phrases. Tey represent nouns and
any words that depend on and accompany nouns. For example, in the sentence The big
red apple fell on the scared cat, the noun chunks are the big red apple and the scared cat.
Extracting these noun chunks is instrumental to many other downstream NLP tasks, such
as named entity recognition and processing entities and relationships between them. In
this recipe, we will explore how to extract named entities from a piece of text.

IMPORT SPACY PACKAGE  

```python
import spacy
```

READ IN THE TEXT FILE

```python
with open(r"/content/001_Study_in_Scarlet.txt", "r", encoding = "utf-8") as file:
  filename = file.read()
```

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp(filename)
```

THE NOUN CHUNKS ARE CONTAINED IN THE DOC.NOUN_CHUNKS CLASS VARIABLE.

```python
for noun_chunk in doc.noun_chunks:
  print(noun_chunk.text)
```
