IMPORT LIBRARIES

```python
import spacy
from spacy import displacy
from pathlib import Path
```

LOAD THE spacy ENGINE

```python
nlp = spacy.load("en_core_web_sm")
```

DEFINE THE VISUALIZE FUNCTION, WHICH WILL CREATE THE DEPENCY VISUALIZATION

```python
def visualize(doc, is_list = False):
  options = {"add_lemma": True,
             "compact": True,
             "color": "green",
             "collapse_punct": True,
             "arrow_spacing": 20,
             "bg": "#FFFFE6",
             "font": "Times",
             "distance": 120
             }
  if (is_list):
    displacy.serve(list(doc.sents), style = "dep",
                   options = options)
  else:
    displacy.serve(doc, style = "dep", options = options)
```

DEFINE A SHORT TEXT FOR PROCESSING

```python
short_text = "The great diversity of life today evloved from less-diverse ancestral organisms over billions of years"
```

CREATE A DOC OBJECT AND PROCESS IT USING THE VISUALIZE FUNCTION:

```python
doc = nlp(short_text)
visualize(doc)
```
