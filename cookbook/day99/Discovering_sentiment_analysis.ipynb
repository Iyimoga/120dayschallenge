# How to do itâ€¦
We will defne two functions: one will do sentiment analysis using NLTK, and the other
using TextBlob.

  import packages

```python
from textblob import TextBlob
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

Define the Sentiment List

```python
sentences = ["I love going to school!", "I hate going to school!"]
```

Initialize the nltk sentiment engine

```python
sid = SentimentIntensityAnalyzer()
```

Define the get_blob_sentiment function, which will use the textblob pachage to determine the sentence sentiment

```python
def get_blob_sentiment(sentence):
  result = TextBlob(sentence).sentiment
  print(sentence,result.polarity)
  return result.polarity
```

Define the get_nltk_sentiment function, which will use the nltkSentimentIntensityAnalyzer:

```python
def get_nltk_sentiment(sentence):
  ss = sid.polarity_scores(sentence)
  print(sentence, ss["compound"])
  return ss["compound"]
```

first, use the NLTK function on each sentence in the lsit:

```python
for sentence in sentences:
  sentiment = get_nltk_sentiment(sentence)
```

Next, use the TextBlob function:


```python
for sentence in sentences:
  sentiment = get_blob_sentiment(sentence)
```
