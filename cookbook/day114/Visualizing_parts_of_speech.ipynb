IMPORT PACKAGES

```python
import spacy
import nltk
nltk.download("punkt")
from nltk.tokenize import sent_tokenize
import matplotlib.pyplot as plt
```

LOAD spacy ENGINE

```python
nlp = spacy.load("en_core_web_sm")
```

```python
past_tags = ["VBD", "VBN"]
present_tags = ["VBP", "VBZ", "VBG"]
```

LOAD TEXT

```python
with open(r"/content/003_ASH_01_Scandal_In_Bohemia.txt", "r", encoding = "utf-8") as file:

  text = file.read()
```

```python
sentences = sent_tokenize(text)
```

```python
for sentence in sentences:
  print(sentence)
```

DEFINE THE VISUALIZE_VERBS FUNCTION, WHICH WILL CREATE THE VISUALIZATION

```python
def visualize_verbs(sentences):
    doc = nlp(" ".join(sentences))

    verb_dict = {"Inf": 0, "Past": 0, "Present": 0}

    for token in doc:
        if token.tag_ == "VB":
            verb_dict["Inf"] += 1
        elif token.tag_ in past_tags:
            verb_dict["Past"] += 1
        elif token.tag_ in present_tags:
            verb_dict["Present"] += 1

    plt.figure(figsize=(12,6))
    plt.bar(verb_dict.keys(), verb_dict.values())
    plt.title("Verb Tense Distribution")
    plt.xlabel("Verb Type")
    plt.ylabel("Count")
    plt.show()
```

```python
visualize_verbs(sentences)
```
