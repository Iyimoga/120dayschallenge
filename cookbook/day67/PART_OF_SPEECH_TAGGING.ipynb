# PART OF SPEECH TAGGING(USING SPACY)

In many cases, NLP processing depends on determining the parts of speech of the words
in the text. For example, in sentence classifcation, we sometimes use the parts of speech
of the words as a feature that is input to the classifer. In this recipe, we will again consider
the NLTK and spaCy algorithms.

In this recipe, we will use the spaCy package to label words with their parts of speech, and
I will show that it is superior to NLTK in this task.

IMPORT SPACY PACKAGE

```python
import spacy
```

READ IN THE BOOK TEXT

```python
filename = "002_Sign_of_Four.txt"
file = open(r"/content/002_Sign_of_Four.txt", "r", encoding = "utf-8")
text = file.read()
```

REPLACE NEWLINES WITH SPACES

```python
text = text.replace("\n", " ")
```

INITIALIZE THE SPACY ENGINE

```python
nlp = spacy.load("en_core_web_sm")
```

USE SPACY ENGINE TO PROCESS THE TEXT

```python
doc = nlp(text)
```

GET THE LIST OF TUPLES WITH WORDS AND THE PART OF SPEECH TAGS:

```python
words = [token.text for token in doc]
pos = [token.pos_ for token in doc]
word_pos_tuples = list(zip(words, pos))
```

```python
print(word_pos_tuples)
```

# PART OF SPEECH TAGGING(USING NLTK)

IMPORT THE NLTK PACKAGE

```python
import nltk
```

READ IN THE BOOK TEXT

```python
filename = "002_Sign_of_Four.txt"
file = open(r"/content/002_Sign_of_Four.txt", "r", encoding = "utf-8")
text = file.read()
```

REPLACE NEWLINES WITH SPACES

```python
text = text.replace("\n", " ")
```

TOKENIZE THE TEXT IN WORDS

```python
import nltk
nltk.download("punkt_tab")
words = nltk.tokenize.word_tokenize(text)
```

PROCESS THE LIST OF WORDS WITH THE NLTK PARTS OF SPEECH TAGGER

```python
nltk.download('averaged_perceptron_tagger_eng')
words_with_pos = nltk.pos_tag(words)
```

```python
print(words_with_pos)
```
