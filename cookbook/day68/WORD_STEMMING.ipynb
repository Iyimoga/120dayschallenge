# WORD STEMMING

In some NLP tasks, we need to stem words, or remove the sufxes and endings such as
-ing and -ed. This recipe shows how to do that.

IMPORT NLTK SNOWBALL STEMMER

```python
from nltk.stem.snowball import SnowballStemmer
```

INITIALIZE STEMMER WITH ENGLISH

```python
stemmer  = SnowballStemmer("english")
```

INITIALIZE A LIST WITH WORDS TO STEM:

```python
words = ["leaf", "leaves", "booking", "writing",
         "completed", "stemming", "skies"]
```

STEM THE WORDS

```python
stemmed_words = [stemmer.stem(word) for word in words]
```

```python
print(stemmed_words)
```
