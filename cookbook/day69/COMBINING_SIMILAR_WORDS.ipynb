# COMBINING SIMILAR WORDS

A similar technique to stemming is lemmatization. Te diï¬€erence is that lemmatization
provides us with a real word, that is, its canonical form. For example, the lemma of the
word cats is cat, and the lemma for the word ran is run

Import the nltk WorNet lemmertizer

```python
from nltk.stem import WordNetLemmatizer
```

Initialize lemmatizer

```python
lemmatizer = WordNetLemmatizer()
```

Initialize a list with words to lemmatize

```python
words = ["duck", "geese", "cats", "books"]
```

Lemmatize the words

```python
import nltk
nltk.download("wordnet")
lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
```

```python
print(lemmatized_words)
```
