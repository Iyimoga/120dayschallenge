# Loading a Dataset

In this recipe, we will learn how to load a public dataset and work with it. We will use the RottenTomatoes dataset for this recipe as an example. This dataset contains ratings and reviews for movies.

### How to do it...

In this recipe, you will load the RottenTomatoes dataset from the HuggingFace site using the dataset package. This package will download the dataset for you if it does not exist.

Import necessary types and functions from the datasets package

```python
from datasets import load_dataset, get_dataset_split_names
```

Load "rotten tomatoes" via the load_dataset function and print the internal dataset splits. This dataset contains the train, validation, and test splits:

```python
dataset = load_dataset("rotten_tomatoes")
print(get_dataset_split_names("rotten_tomatoes"))
```

Load the dataset and print the attributes of the train split. The training_data.description describes the dataset details and the training_data.features describes the features of the dataset. In the output, we can see that the training_data split contains the features text, which is of the type string, and label, which is of type categorical, with the values neg and pos:

```python
training_data = dataset["train"]
print(training_data.description)
print(training_data.features)
```

Now that we have loaded the dataset, we will print the first five sentences from it. This is just to confirm that we are indeed able to read from the dataset:

```python
sentences = training_data["text"][:5]
[print(sentence) for sentence in sentences]
```
